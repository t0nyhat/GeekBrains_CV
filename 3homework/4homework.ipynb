```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. Задание (в программе)
# Нарисуйте график функции:
# y(x) = k∙cos(x – a) + b
# для некоторых (2-3 различных) значений параметров k, a, b

x = np.linspace(-5*np.pi, 5*np.pi, 201)
k = 1
a = 1
b = 1
k1 = 4
a1 = 5
b1 = 5
plt.plot(x, k*np.cos(x - a) + b)
plt.plot(x, k1*np.cos(x - a) + b)
plt.plot(x, k*np.cos(x - a1) + b)
plt.plot(x, k*np.cos(x - a) + b1)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
