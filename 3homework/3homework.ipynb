```python
"""
1. Задание
Даны два вектора в трехмерном пространстве: (10,10,10) и (0,0,-10)
Найдите их сумму. (на листочке)
Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)

Ответ: сумма векторов (10,10,0)

"""
v = [10, 10, 0]
a = math.sqrt(math.pow(v[0],2) + math.pow(v[1],2) + math.pow(v[2],2))
print("Длина вектора равна - ",a)
```

```python
"""
2. Задание
Почему прямые не кажутся перпендикулярными? 


Ответ: из-за разного масштаба по осям

"""

x = np.linspace(-5, 5, 21)
y = 3*x+1
y2 = (-1/3)*x+1
#plt.axis('equal')
plt.plot(x,y)
plt.plot(x,y2)
plt.xlabel("x")
plt.ylabel("y")
```

```python
# 3. Задание (в программе)
# окружность
x = []
y1 = []
y2 = []
r = 10
for i in range(-100,101):
    x_ = i/100 * r
    x.append(x_)
    y1.append(math.sqrt(math.pow(r,2)-math.pow(x_,2)))
    y2.append(-math.sqrt(math.pow(r,2)-math.pow(x_,2)))
plt.axis('equal')
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel("x")
plt.ylabel("y")
```

```python
# 3. Задание (в программе)
# овал
x = []
y1 = []
y2 = []
r = 10
r2 = 2
for i in range(-100,101):
    x_ = i/100 * r
    x.append(x_)
    y1.append(math.sqrt(math.pow(r,2)-math.pow(x_,2))/r2)
    y2.append(-math.sqrt(math.pow(r,2)-math.pow(x_,2))/r2)
plt.axis('equal')
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel("x")
plt.ylabel("y")
```

```python
# 3. Задание (в программе)
# гипербола
x = []
y1 = []
y2 = []
r = 10
r2 = 2
for i in range(-100,101):
    x_ = i/100 * r
    x.append(x_)
    y1.append(math.sqrt(math.pow(r,2)+math.pow(x_,2))/r2)
    y2.append(-math.sqrt(math.pow(r,2)+math.pow(x_,2))/r2)
plt.axis('equal')
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel("x")
plt.ylabel("y")
```

```python
# 5. Задание (в программе)
# Нарисуйте трехмерный график двух параллельных плоскостей.

from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 2)
Y = np.arange(-50, 50, 20)
X, Y = np.meshgrid(X, Y)
Z = X*2 + Y*2
Z2 = X*2 + Y*2 + 100
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Z2)
```

```python
# 5. Задание (в программе)
# Нарисуйте трехмерный график двух любых поверхностей второго порядка.

from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z = X**2 + Y**2 
Z2 = X**2 - Y**2
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Z2)
```
