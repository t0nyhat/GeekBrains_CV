```python
#5. Дополните код расчетом коэффициента корреляции x и y по формуле

n = 100
r = 0.7
x = np.random.rand(n)
y = r*x + (1 - r)*np.random.rand(n)
plt.plot(x, y, 'o')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

a = (np.sum(x)*np.sum(y) - n*np.sum(x*y))/(np.sum(x)*np.sum(x) - n*np.sum(x*x))
b = (np.sum(y) - a*np.sum(x))/n



A = np.vstack([x, np.ones(len(x))]).T
a1, b1 = np.linalg.lstsq(A, y)[0]
print(a, b)
print(a1, b1)
plt.plot([0, 1], [b, a + b])
plt.show()

c = np.corrcoef(x, y) 

cc = np.sum((x-np.average(x))*(y-np.average(y)))/np.sqrt( np.sum((x-np.average(x))*(x-np.average(x)))*np.sum((y-np.average(y))*(y-np.average(y))) )

print(c)
print("Коэффициент корреляции - ",cc)
```
